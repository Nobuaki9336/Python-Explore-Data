#### Filtering and Ordering in Pandas 

```python
# Filter BMW Car 
df[df['Car Make'] == 'BMW']
```

```python
df['Price (in USD)'] = df['Price (in USD)'].str.replace(',', '')
df['Price (in USD)'] = pd.to_numeric(df['Price (in USD)'])
```

```python
df[df['Price (in USD)'] > 2500000]
```

```python
# Remove ',' & '+' sign in the column and then change the dtype to numeric 
df['Horsepower'] = df['Horsepower'].str.replace(r'[+,]', '', regex=True)
df['Horsepower'] = pd.to_numeric(df['Horsepower'])
```

```python
# Filer Car Horsepower over or equal 1000 and order it in dscending order
df[df['Horsepower'] >= 1000].sort_values(by='Horsepower',ascending=False)
```

```python
columns_to_keep = ['Car Make', 'Horsepower','Price (in USD)']

df2 = df[columns_to_keep]

df2
```

### Questions

```python
# What are the most common crimes
df.OFFENSE_CODE_GROUP.value_counts()
```

```python
offense_group_vals = df.OFFENSE_CODE_GROUP.value_counts()[:10]

display((offense_group_vals / df.shape[0])* 100)
```

```python
# Creating a bar chart of the Top 10 offense groups  

plt.figure(figsize=(9, 5))

((offense_group_vals / df.shape[0])* 100).plot(kind='bar');
plt.title('Top 10 Offense Groups');
```

```python
# What are the least common offense groups?
df.OFFENSE_CODE_GROUP.value_counts().sort_values(ascending=True)[:10]
```
